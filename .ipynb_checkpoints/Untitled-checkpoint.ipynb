### **Merging images from all the four magnification folder**

```python
import shutil

# Base folder with magnifications
base_data_dir = "data/dataset_cancer_v1/classificacao_binaria"
# Destination folder
merged_dir = "data/classificacao_binaria_merged"
os.makedirs(merged_dir, exist_ok=True)

# Classes
classes = ["benign", "malignant"]
for cls in classes:
    os.makedirs(os.path.join(merged_dir, cls), exist_ok=True)

# Loop through magnifications and copy images
for mag in ["40X", "100X", "200X", "400X"]:
    for cls in classes:
        cls_path = os.path.join(base_data_dir, mag, cls)
        for img in os.listdir(cls_path):
            src = os.path.join(cls_path, img)
            dst = os.path.join(merged_dir, cls, img)
            # Rename if duplicate
            if os.path.exists(dst):
                name, ext = os.path.splitext(img)
                dst = os.path.join(merged_dir, cls, f"{name}_{mag}{ext}")
            shutil.copy2(src, dst)

print("All images merged into:", merged_dir)
```

### **Checking for duplicates**

```python
def count_duplicates(dataset):
    seen = set()
    duplicates = 0
    total = 0

    for images, _ in dataset:
        for img in images:
            img_hash = hashlib.md5(img.numpy().tobytes()).hexdigest()
            if img_hash in seen:
                duplicates += 1
            else:
                seen.add(img_hash)
            total += 1
        

    print(f"\nTotal images checked: {total}")
    print(f"Unique images: {len(seen)}")
    print(f"Duplicate images: {duplicates}")

# Run the check
count_duplicates(data)
```

### **Checking for Corrupt Images**

```python
# Path of the dataset
dataset_path = "data/classificacao_binaria_merged"
classes = ['benign', 'malignant']

# Initialize list to store corrupted images
corrupted_images = []

for c in classes:
    class_path = os.path.join(dataset_path, c)
    for img_name in os.listdir(class_path):
        img_path = os.path.join(class_path, img_name)
        try:
            img = Image.open(img_path)
            img.verify()  # Verifys that image is not corrupted
        except Exception:
            corrupted_images.append(img_path)

print(f"Total corrupted images found: {len(corrupted_images)}")
if corrupted_images:
    print("Some images are corrupted. Example:")
    print(corrupted_images[:5])
else:
    print("No corrupted images found.")
```

### **Remove duplicated images**

```python
def remove_duplicates(base_data_dir, clean_data_dir, classes=['benign', 'malignant']):
    os.makedirs(clean_data_dir, exist_ok=True)
    seen = set()

    for cls in classes:
        cls_clean_path = os.path.join(clean_data_dir, cls)
        os.makedirs(cls_clean_path, exist_ok=True)
        cls_path = os.path.join(base_data_dir, cls)

        for img_name in os.listdir(cls_path):
            img_path = os.path.join(cls_path, img_name)
            
            try:
                img = Image.open(img_path)
                img_hash = hashlib.md5(img.tobytes()).hexdigest()
                img.close()

                if img_hash not in seen:
                    seen.add(img_hash)
                    shutil.copy2(img_path, os.path.join(cls_clean_path, img_name))

            except Exception as e:
                print(f"Skipping corrupted image: {img_path} — {e}")

    print(f"Duplicate removal complete! Clean dataset saved at: {clean_data_dir}")


# Usage
#remove_duplicates("data/classificacao_binaria_merged", "data/classificacao_binaria_clean")
```

## **MODELING**

### **Convolutional Neaural Network for binary classification**

```python
# CNN Model
model = Sequential([
    
    data_augmentation,                                           

    # rescale pixels
    layers.Rescaling(1./255, input_shape=(224, 224, 3)),         # Scaling corrects normalization for stability

    # Convolutional Block 1
    layers.Conv2D(32, (3,3), activation="relu", kernel_initializer='he_normal'),     # avoids vanishing gradients
    layers.MaxPooling2D(),                                  

    # Convolutional Block 2
    layers.Conv2D(64, (3,3), activation="relu", kernel_initializer='he_normal'), 
    layers.MaxPooling2D(),

    # Convolutional Block 3
    layers.Conv2D(128, (3,3), activation="relu", kernel_initializer='he_normal'), 
    layers.MaxPooling2D(),

    layers.GlobalAveragePooling2D(),                           # Reduces parameters, Prevents overfitting, More stable training

    # dense layers
    layers.Dense(128, activation='relu'),
    layers.Dropout(0.5),                                       # Regularization reduces overfitting

    # Output layer for binary classification
    layers.Dense(1, activation='sigmoid')                       # Sigmoid Proper for binary classification
])

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=1e-4),  # slightly lower LearningRate for stability/ Adam optimizer for fast convergence
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Model summary
model.summary()
```

```python
# Callbacks
# stops training automatically when the model stops improving
early_stop = EarlyStopping(
    monitor='val_loss',
    patience=5,
    restore_best_weights=True
)

# Lowers learning rate when training stops improving
reduce_lr = ReduceLROnPlateau(                        
    monitor='val_loss',
    factor=0.3,
    patience=3,
    min_lr=1e-7
)
# Save the best version of your model during training
checkpoint = ModelCheckpoint(                       
    "best_cnn_model.keras",
    monitor="val_accuracy",
    save_best_only=True
)

# Train the model
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=25,
    class_weight=class_weights,
    callbacks=[early_stop, class_weight=class_weights, reduce_lr, checkpoint]
)
```

* Setting up the training data using an ImageDataGenerator with augmentation ensures that the model sees slightly different versions of each image in every epoch. This prevents the CNN from simply memorizing the training images and helps it learn strong, generalizable features for accurate classification.

```python

# Define dataset directories
# Base directories
base_data_dir = "data/classificacao_binaria_merged"     


# Prepare the training data with augmentation
train_data=ImageDataGenerator(
    rescale=1./255,                                        # Normalization/ Converts pixel values from 0–255 to 0–1 add more stability
    validation_split=0.2,                        # 20% validation
    rotation_range=20,                                     # Rotates images up to 20° for augmentation
    width_shift_range=0.1,                                 # Randomly shifts images horizontally/vertically by up to 10% to improve generalization
    height_shift_range=0.1,
    zoom_range=0.1,                                        # zooms images slightly for robustness
    horizontal_flip=True,                                  # Flip images to simulate different orientations.
    vertical_flip=True
)
# Create generator to feed images to CNN
train_generator = train_data.flow_from_directory(
     base_data_dir,
     target_size=(224,224),                                   # Resizes all images to 224×224 pixels, standard input size for CNNs
     batch_size=32,                                           # Feeds images in batches of 32 to the CNN
     subset='training',
     class_mode='binary',                                     # Labels are binary (0 = benign, 1 = malignant)
     shuffle=True                                             # Randomizes the order of images each epoch to improve training stability.
)



# Prepare Validation data (only scaling is done for this part)
val_data= ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2)



# Create a generator for validation set
val_generator = val_data.flow_from_directory(
    base_data_dir,
    target_size=(224,224),
    batch_size=32,
    subset='validation',
    class_mode='binary',
    shuffle=True)
```

```python
from tensorflow.keras.preprocessing.image import array_to_img

# Get a batch of images and labels
images, labels = next(train_generator)

# Show first 6 images in the batch
plt.figure(figsize=(10, 6))
for i in range(6):
    plt.subplot(2, 3, i + 1)
    plt.imshow(array_to_img(images[i]))
    label = "Malignant" if labels[i] == 1 else "Benign"
    plt.title(label)
    plt.axis("off")

plt.show()
```

```python
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import GlobalAveragePooling2D, Dense, Dropout
from tensorflow.keras.applications import ResNet50
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint

# Image parameters
IMG_SIZE = (224, 224)
BATCH_SIZE = 32

# Load ResNet50 without top layers, use ImageNet weights
base_model = ResNet50(weights='imagenet', include_top=False, input_shape=(IMG_SIZE[0], IMG_SIZE[1], 3))

# Freeze the convolutional base
base_model.trainable = False

# Add custom classifier head
x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(512, activation='relu')(x)
x = Dropout(0.5)(x)
output = Dense(1, activation='sigmoid')(x)  # binary classification

# Define the model
model = Model(inputs=base_model.input, outputs=output)

# Compile the model
model.compile(
    optimizer=Adam(learning_rate=1e-4),
    loss='binary_crossentropy',
    metrics=['accuracy']
)

# Callbacks
early_stop = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
checkpoint = ModelCheckpoint('resnet50_bc_model.keras', monitor='val_loss', save_best_only=True)

# Train the model
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=20,
    callbacks=[early_stop, checkpoint]
)
```

```python
# Plot accuracy
plt.plot(history.history['accuracy'], label='train_accuracy')
plt.plot(history.history['val_accuracy'], label='val_accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Plot loss
plt.plot(history.history['loss'], label='train_loss')
plt.plot(history.history['val_loss'], label='val_loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```
