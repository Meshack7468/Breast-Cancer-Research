# BUSINESS UNDERSTANDING

## OVERVIEW
Breast cancer remains a leading illness affecting women worldwide, However, it’s not a single disease, it consists of several biologically distinct types such as Luminal A, Luminal B, HER2-enriched, Basal-like. These subtypes are usually identified using a genomic test called PAM50 Assay, which helps understand how aggressive a tumor is and determine the most effective treatment plan. Unfortunately, such tests are not always easily accessible.
This project uses machine learning to predict patient’s tumor molecular subtype and patients survival status using machine learning. By doing so, it offers a data-driven tool that can make better treatment recommendations and prognosis decisions. 

## PROBLEM STATEMENT
Different breast cancer  subtypes vary in tumor aggressiveness and treatment strategy. These subtypes are usually identified using an advanced test, which is not always easily accessible. Without this test there is limited insight into tumor behavior and the most effective treatment approach. 
Many patients also often seek clear information about their survival outcomes, yet survival predictions rely on multiple clinical and genomic factors that require deep analytical skills.

This project addresses these challenges by accurately predicting a patient's breast cancer molecular subtype and survival outcome using clinical data. By providing these insights, it supports informed treatment decisions and aims to improve patient survival outcomes.


## BUSINESS OBJECTIVES
This project has three main goals that connect machine learning predictions to real-world clinical value:

1. **Molecular Subtype Prediction** — Build a model that predicts the PAM50 + Claudin-low molecular subtypes of breast cancer (e.g., Luminal A,
Luminal B, HER2-enriched, Basal-like, and Claudin-low).
This model can act as a digital alternative to the molecular PAM50 genomic test, helping in understanding tumor biology and guiding personalized treatment decisions.
3. **Binary Survival Prediction** — Build a model that predicts whether a patient is likely to survive or not (LIVING or DECEASED), to help identify patients who are at high risk early and plan closer monitoring.

5. **Multi-class Vital Status Prediction** — Build a model that further classifies the survival state of patients as Died of Disease, or Died of Other Causes, or living to distinguish between cancer-related and non-cancer death. 

6. Deploy the models into a user-friendly web application that ensures the project is accessible, practical, and usable in real world healthcare settings.

# DATA UNDERSTANDING

* The dataset is from the METABRIC (Molecular Taxonomy of Breast Cancer International Consortium) study — a comprehensive breast cancer dataset widely used for clinical and genomic research.
* It contains data from 2,509 patients,  Each row represents a unique patient, and each column captures a clinical, diagnostic, or treatment-related attribute.
  
**Structure of the Data**
* Total records (rows) are 2,509, Total features (columns) are 39
* Data types our data contain 12 numerical columns & 27 categorical columns
* Dataset is in a tsv file format (Tab-separated values)

**Key Features**
* Some key columns include:
1. Pam50 + Claudin-low subtype - Molecular subtype classification (Luminal A, Luminal B, HER2-enriched, Basal-like, Normal-like). Critical for linking molecular biology to treatment response and prognosis
2. Nottingham Prognostic Index (NPI) - A single summary score that combines tumor size, grade, lymph node status — widely used in clinical prognosis prediction.
3. Age at Diagnosis	Age (in years) - when the patient was diagnosed with breast cancer.
4. Type of Breast Surgery -	Indicates the type of surgical procedure performed (e.g., mastectomy, breast-conserving).
5. Inferred Menopausal State - Derived feature (Pre/Post-menopausal) it influences hormone receptor status and therapy choice.
6. Lymph nodes examined positive - Indicates number of lymph nodes with metastasis 
7. Chemotherapy / Hormone Therapy / Radio Therapy -	Binary treatment indicators whether a patient received each therapy.
8. ER, PR, HER2 Status	Biomarkers - indicating estrogen, progesterone, and HER2 receptor status — critical for determining cancer subtype and treatment plan.
9. Neoplasm Histologic Grade - Describes how abnormal a tumor cells look under a microscope & how fast they are likely to grow and spread
10. Tumor Stage / Tumor Size -	Describe the aggressiveness, spread, and physical characteristics of the tumor.
11. Overall Survival (Months) / Status - Time and outcome of patient survival data — useful for prognosis analysis.
12. Overall Survival Status / Vital Status	Indicates whether the patient is alive or deceased.
13. Relapse Free Status (Months) / Status -	Time until recurrence or relapse, useful for recurrence prediction.
14. TMB (Tumor Mutation Burden)	- Number of nonsynonymous mutations per sample — a genomic feature.
15. Study ID / Patient ID / Sample ID	- Unique identifiers for the study and individual patients.
16. Cellularity -  Refers to how many of the cells in a tissue sample are cancer cells compared to normal cells.
17. Cancer Type / Cancer Type Detailed - Describes the general and specific form of breast cancer.

**Data Quality and Completeness**

* Several columns contain missing values: i.e
1. Type of Breast Surgery → 554 values missing
2. Cellularity → 592 values missing
3. Tumor Stage → 721  missing
4. 3-Gene classifier subtype → 745 missing
5. Overall Survival and related fields → ~528 missing
* Since most of the nans are less than half of the data in each column, we will fill nans in numerical columns with median and  drop nans in categorical columns


**Target Variables**
1. "Pam50 + Claudin-low subtype" It's Multi-class used in Molecular Subtype Prediction. Predicts tumor molecular subtype (Luminal A, Luminal B, HER2-enriched, Basal-like, Normal-like).
2. "Overall Survival Status" Used in Binary Survival Prediction to predict if the patient survived (LIVING) or not (DECEASED).
3. "Patient’s Vital Status"	Multi-class	used in vital Outcome Prediction it classifies patients as Living, Died of Disease, or Died of Other Causes.

```python
# Drop irrelevant and redundant columns
drop_cols = [
    "Study ID", "Patient ID", "Sample ID",
    "HER2 status measured by SNP6", "ER status measured by IHC",
    "Number of Samples Per Patient", "Cancer Type"
]
df = df.drop(columns=drop_cols)
```

```python
# separeted numerical columns and categorical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns
```

```python
# Fill the Nans in categorical columns with mode
df[categorical_cols] = df[categorical_cols].fillna(df[categorical_cols].mode().iloc[0])

# Fill nans in numeric columns with median
df[numerical_cols] = df[numerical_cols].fillna(df[numerical_cols].median())
```

```python
# Remove duplicates
df = df.drop_duplicates()
```

# **Modeling**

### a)  **Molecular subtype prediction**

* **Goal:** Classify tumors into molecular subtypes using clinical and genetic features.

By predicting these subtypes our model will:
1. Help doctors estimate the tumor’s molecular subtype without needing the actual PAM50 genetic test.
2. Give insights into how aggressive or treatable a tumor might be.
3. Guide treatment decisions.


```python
# Import relevant libraries
from sklearn.preprocessing import OneHotEncoder, StandardScaler, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestClassifier
from imblearn.pipeline import Pipeline as ImbPipeline
from imblearn.over_sampling import SMOTE
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pickle
from sklearn.linear_model import LogisticRegression
from imblearn.pipeline import Pipeline as ImbPipeline
from xgboost import XGBClassifier
from sklearn.decomposition import PCA
from sklearn.preprocessing import FunctionTransformer
from imblearn.over_sampling import SMOTENC
```

```python
# Check the distribution of the target column "Pam50 + Claudin-low subtype"

# Visualize the class distribution of "Pam50 + Claudin-low subtype"
df["Pam50 + Claudin-low subtype"].value_counts().plot(kind='bar', figsize=(6,4))

plt.title("Distribution of PAM50 Subtypes")
plt.xlabel("Subtype")
plt.ylabel("Count")
plt.show()
```

```python
# Select the subtypes to predict
df = df[df["Pam50 + Claudin-low subtype"].isin(["LUMA","LUMB","CLAUDIN-LOW","HER2","BASAL"])]

# Check the counts
df["Pam50 + Claudin-low subtype"].value_counts()
```

```python
# target and features 

# Select only the desired feature columns to use
feature_cols = [
    'Age at Diagnosis',
    'Type of Breast Surgery',
    'ER Status',
    'HER2 Status',
    'TMB (nonsynonymous)',
    'Tumor Stage',
    '3-Gene classifier subtype',
    'PR Status',
    'Lymph nodes examined positive',
    'Integrative Cluster',
    'Nottingham prognostic index',
    'Tumor Other Histologic Subtype',
]

# Target column
y = df['Pam50 + Claudin-low subtype']
# Features dataframe
X = df[feature_cols]
```

```python
# LabelEncode the target column because it is a multi-class classification
le = LabelEncoder()
y_encoded = le.fit_transform(y)

# Save the LabelEncoder for later use in Streamlit
#joblib.dump(le, "models/molecular_le.joblib")
with open("models/molecular_le.pkl", "wb") as file:
    pickle.dump(le, file)

print("LabelEncoder saved successfully as molecular_le.pkl")
```

```python
# Train-test split. We use stratify to balance the classes in both training & testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

```python
# Filter columns
categorical_columns = [col for col in X_train.select_dtypes(include='object').columns]
numerical_columns = [col for col in X_train.select_dtypes(include=['int64', 'float64']).columns]
```

```python
# One-Hot Encode categorical features & standard scale numerical columns
preprocessor = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_columns),
        ('num', StandardScaler(), numerical_columns)
])
```

```python


# We must pass SMOTENC the indices of categorical columns
cat_indices = [X_train.columns.get_loc(col) for col in categorical_columns]


# Build pipeline
pipelineRF =ImbPipeline(steps=[
    ("preprocess", preprocessor),
    ("smote", SMOTENC(random_state=42, categorical_features=cat_indices)),
    ("classifier", RandomForestClassifier(
        random_state=42,
        class_weight="balanced",                                           # help with class imbalance especially for rare subtypes
        n_jobs=-1
        
    ))
])


# Hyperparameter grid for Random Forest
param_grid = {
    'classifier__n_estimators': [100, 150, 200, 300],            # Total number of trees in ensemble
    'classifier__max_depth': [ 5, 10, None],                     # max depth of the trees
    'classifier__min_samples_split': [2, 5, 8],
    'classifier__min_samples_leaf': [1, 2, 4],
    
}


# GridSearchCV with 3-fold cross-validation
grid_search_RF = GridSearchCV(estimator=pipelineRF, param_grid=param_grid, cv=3, scoring='f1_macro', n_jobs=-1)

# Fit GridSearch
grid_search_RF.fit(X_train, y_train)

# Get the best model
best_RF = grid_search_RF.best_estimator_
print("Best Parameters:", grid_search_RF.best_params_)

# save model
with open("molecular_subtype_model.pkl", "wb") as file:
    pickle.dump(best_RF, file)

print("Pipeline saved successfully as molecular_subtype_model.pkl")
```

```python
# Evaluation

# Predict on test set
y_pred = grid_search_RF.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred, target_names=le.classes_))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(8,6))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le.classes_, yticklabels=le.classes_, cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title(" Random Forest — Molecular subtype prediction Confusion Matrix")
#plt.savefig("Molecular subtype prediction cm.png")
plt.show()
```

* **The random forest model correctly captures 200 Luminal A patients, 67 Luminal B, 20 HER2, 18 Claudin-low and 25 Basal patients.**

#### **Random Forest Molecular subtype prediction Feature importance**

```python
#  Extract classifier
rf_model = best_RF.named_steps['classifier']
importances = rf_model.feature_importances_

# Extract categorical feature names from preprocessor
preprocessor = best_RF.named_steps['preprocess']
ohe = preprocessor.named_transformers_['cat']
cat_features = ohe.get_feature_names_out()

# Combine with numeric feature names
all_features = list(cat_features) + numerical_columns

# Create DataFrame
feature_importance_RF = pd.DataFrame({
    "Feature": all_features,
    "Importance": importances
}).sort_values(by="Importance", ascending=False)

# Top 15 features
feature_importance_RF.head(15)
```

* The top most features that influence Pam50 + Claudin-low subtypes prediction for Random Forest are 3-Gene Classifier Subtype, ER Status,  Integrative Cluster, Tumor Mutation Burden (TMB),HER2 Status
* **Lets build a XGBoost model and see if the performance increases.**

```python

# XGBoost Pipeline
pipelineXGB = ImbPipeline(steps=[
    ('preprocess', preprocessor),
    ('smote', SMOTENC(categorical_features=cat_indices, random_state=42)),
    ("classifier", XGBClassifier(
        objective="multi:softprob",                # tells xgb this is a multi-class problem
        #num_class=len(np.unique(y_encoded)),       
        eval_metric="mlogloss",                    # optimazation to minimize loss             
        random_state=42,                           # Ensures reproducibility
        n_jobs=-1,                                  # speeds up training
        reg_lambda=1                             # L2 penalty/Ridge prevents overfitting & improve model stability
    ))
])

# Hyperparameter grid
paramX_grid = {
    'classifier__n_estimators': [100, 200, 300],        # Total number of trees in ensemble
    'classifier__max_depth': [3, 5],                    # max depth of each tree
    'classifier__learning_rate': [0.01, 0.1],           # how fast the model learns
    'classifier__subsample': [0.8],                     # prevents overfitting
    'classifier__colsample_bytree': [0.8]               # improves generalization
}

# GridSearchCV
grid_search_XGB = GridSearchCV(
    estimator=pipelineXGB, 
    param_grid=paramX_grid, 
    cv=3,                                       # ensure model is validated on different subset of training
    scoring='f1_macro',                         # balances performance across all classes
    n_jobs=-1,)

# Fit GridSearch
grid_search_XGB.fit(X_train, y_train)

# Get the best model
best_XGB = grid_search_XGB.best_estimator_
print("Best Parameters:", grid_search_XGB.best_params_)


# Predict
y_pred = grid_search_XGB.predict(X_test)

# Evaluate
print("Accuracy:", accuracy_score(y_test, y_pred))
print(classification_report(y_test, y_pred, target_names=le.classes_))
```

* Build a model that Predict whether a patient is **alive or deceased.** it will:
1. Help identify high-risk patients who may need closer monitoring or aggressive therapy.
2. Provide insights into prognostic factors associated with survival.
3. Assist in treatment planning and patient counselling.

```python
# Create feature and target DataFrames

# Target variable
y_binary = df["Overall Survival Status"]

# Define the selected feature columns
feature_cols_binary = [
    'Age at Diagnosis',
    'Type of Breast Surgery',
    'ER Status',
    'HER2 Status',
    'TMB (nonsynonymous)',
    'Tumor Stage',
    '3-Gene classifier subtype',
    'PR Status',
    'Lymph nodes examined positive',
    'Integrative Cluster',
    'Hormone Therapy',
    'Nottingham prognostic index',
    
]

# Feature dataframe
X_binary = df[feature_cols_binary]

```

```python
# Used regex to extract numeric part from target strings ("1:DECEASED", 0:LIVING)
y_binary = y_binary.str.extract(r'(\d+)').astype(int)[0]

```

```python
# train test split in test size of 80
X_binary_train, X_binary_test, y_binary_train, y_binary_test = train_test_split(X_binary, y_binary, test_size=0.2, stratify=y_binary, random_state=42)
    
```

```python
# Filter categorical and numerical columns
categorical_cols = [col for col in X_binary_train.select_dtypes(include='object').columns]
numerical_cols   = [col for col in X_binary_train.select_dtypes(include=['int64', 'float64']).columns]
```

```python
# One-Hot Encode categorical features & standard scale categorical columns
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), categorical_cols),
        ("num", StandardScaler(), numerical_cols)])
```

```python

# SMOTE for class balancing
smote = SMOTE(random_state=42)

# Define the pipeline
pipelinelr= ImbPipeline([
    ("preprocessor", preprocessor),
    ("smote", smote),
    ("classifier", LogisticRegression(
        max_iter=1000,                             # Ensures convergance
        class_weight="balanced",                   # Account for class imbalance
        random_state=42))                          
])


# Tuning hyperparameter grid
param_grid = {
    "classifier__penalty": ["l2"],                 # Regularization for stability and reduce overfitting
    "classifier__C": [0.01, 0.1, 1, 10, 100],      # Regularization strength
    "classifier__solver": ["lbfgs", "liblinear"]
}

# GridsearchCV
gridsearch_LR = GridSearchCV(estimator=pipelinelr, param_grid=param_grid, scoring="accuracy", cv=3, n_jobs=-1)

# fit the train data
gridsearch_LR.fit(X_binary_train, y_binary_train)

best_lr = gridsearch_LR.best_estimator_


# Make predictions
y_pred_lr = best_lr.predict(X_binary_test)
y_pred_proba_lr = best_lr.predict_proba(X_binary_test)[:, 1]
```

```python
from sklearn.metrics import (accuracy_score, classification_report,confusion_matrix, ConfusionMatrixDisplay, roc_auc_score)
# Evaluation

# Metrics
print("Accuracy:", accuracy_score(y_binary_test, y_pred_lr))
print("\nClassification Report:\n", classification_report(y_binary_test, y_pred_lr))
print("ROC-AUC:", roc_auc_score(y_binary_test, y_pred_proba_lr))

# Confusion matrix
cm = confusion_matrix(y_binary_test, y_pred_lr)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["LIVING (0)", "DECEASED (1)"])
disp.plot(cmap="Blues")
plt.title("Logistic Regression — Survival Status Prediction Confusion Matrix")
plt.show()
```

* The Logistic Regression correctly captures 238 of DECEASED patients and 107 LIVING patients

### The top important features for  predicting survival status(binary)
